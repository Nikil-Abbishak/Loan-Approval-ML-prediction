Section 5 - Data Cleaning  
Leveraged and reused from Coursework Code Reuse Session 1 / Tutorial 2.

```python
# Drop ID column as it does not contribute to modelling
df = df.drop(columns=['id'])

# Fill missing numerical values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing values in 'age' using mean
mean_age = df['age'].mean()
df['age'] = df['age'].fillna(mean_age)

# Fill missing values in 'loan_interest_rate' using mean
mean_interest = df['loan_interest_rate'].mean()
df['loan_interest_rate'] = df['loan_interest_rate'].fillna(mean_interest)

# Fill missing values in 'payment_default_on_file' using mode
mode_default = df['payment_default_on_file'].mode()[0]
df['payment_default_on_file'] = df['payment_default_on_file'].fillna(mode_default)

# Remove rows with negative loan interest rate values
df = df[df['loan_interest_rate'] >= 0]
```

```python
# Count missing values in each column after data cleaning
df.isnull().sum()
```

Section 6 - Encoding Categorical Variables  
Leveraged and reused from Coursework Code Reuse Session 1 / Tutorial 2.

```python
# Convert categorical variables into numerical using one-hot encoding
df = pd.get_dummies(df, drop_first=True)
```

Section 7 - Preparing Dataset for Classification  
Leveraged and reused from Coursework Code Reuse Session 1 / Tutorial 2.

```python
# Define input features for classification
X_class = df.drop('loan_approval_status', axis=1)

# Define target variable
y_class = df['loan_approval_status']
```

Section 8 - Preparing Dataset for Regression  
Leveraged and reused from Coursework Code Reuse Session 1 / Tutorial 2.

```python
# Filter dataset for approved loans only
df_reg = df[df['loan_approval_status'] == 0]

#drop loan approval status as it isnt needed
df_reg = df_reg.drop(columns=['loan_approval_status'])

# Define regression features
X_reg = df_reg.drop('max_allowed_loan', axis=1)

# Define regression target
y_reg = df_reg['max_allowed_loan']
```
